```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Euclidean distance

```python
v1 = np.array([1,2,3])
v2 = np.array([0,0,0])
```

```python
# numpy broadcasting
v1 + v2 # (1,2,3) + (0,0,0) = (1,2,3) ## vector addition
v2 - v1 # (vector subtraction)
v1 * v2 # elementwise multiplication
v2 / v1 # elementwise division
v1 + 3  # (1,2,3) + 3 = (4,5,6). ## scalar addition
v1 * 3  # (1,2,3) + 3 = (3,6,9). ## scalar multiplication

## universal function
def reverse_num(x):
  return 1/x

# broadcastiong
for x in v1:
  print(reverse_num(x))



1 / v1  # 
```

```python
## v1과 v2에 있는 모든 원소들끼리 뺀 다음에, 제곱하고 다 더해서 루트 씌운거 = Euclidean distance (L2 distance)

distance = np.sqrt(np.sum(np.square(v1-v2))) ## =(v1 -  v2)**2한 값들의 sum한 것들의 루트ㅋㅋ

distance2 = np.linalg.norm(v1-v2) #벡터의 크기 (L2 norm) _ 위의 distance를 정의한 내용과 값과 구현 방법이 일치함. 라이브러리를 쓰는 것 뿐
```
