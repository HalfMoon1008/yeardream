## 결측치 처리

```python
print("* 데이터 타입, 데이터프레임 정보 확인")
print("----------------------------------------------------------------------")
df.info()
print()
print("* 데이터 칼럼별 결측값 확인")
print("----------------------------------------------------------------------")
df.isnull().sum()
```

```python
# 결측치 처리
# State : Deleted - row drop
# Customer_id - row drop
# Course_title - row drop

# coupon_title - 0값
# coupon_discount_amount - 0값
# sale_price,tax_free_amount - 0값, B2B - 1, ETC - 2

# method - col dorp

# Course_id - 고민
```

```python
# 혹시 모를 공백값 NaN 처리
df['state'].replace('', np.nan) #1순위
df['customer_id'].replace('', np.nan)
df['course_title'].replace('', np.nan)

# NaN값 row들 drop 
df.dropna(subset=['state'])
df.dropna(subset=['customer_id'])
df.dropna(subset=['course_title'])

# method - col dorp
df.drop(['method'],axis=1)

# coupon_discount_amount, coupon_title, pg - 0값 처리. titlem, pg은 '0'
df.loc[df['coupon_discount_amount'] != df['coupon_discount_amount'], 'coupon_discount_amount'] = 0
df.loc[df['coupon_title'] != df['coupon_title'], 'coupon_title'] = '0'
df.loc[df['pg'] != df['pg'], 'pg'] = '0'
```
